```python
import matplotlib.pyplot as plt
import numpy as np
print("m.pujitha BU22CSEN0101365")
def visualize_weather_model():
    coefficient_a = 20
    coefficient_b = 9
    constant_c = 7
    x_values = np.linspace(-15, 20, 105)
    y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + constant_c
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label=f"y = {coefficient_a}x + {coefficient_b}x + {constant_c}", color='blue')
    plt.title("Weather Model Visualization - Quadratic Equation")
    plt.xlabel("Input Parameter (x)")
    plt.ylabel("Weather Parameter (y)")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()
visualize_weather_model()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("m.pujitha BU22CSEN0101365")
def visualize_weather_model_input():
    try:
        coefficient_a = float(input("Enter the coefficient a: "))
        coefficient_b = float(input("Enter the coefficient b: "))
        constant_c = float(input("Enter the constant c: "))
        x_values = np.linspace(-15, 20, 105)
        y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + constant_c
        plt.figure(figsize=(8, 6))
        plt.plot(x_values, y_values, label=f"y = {coefficient_a}x + {coefficient_b}x + {constant_c}", color='green')
        plt.title("Weather Model Visualization - User-Defined Quadratic Equation")
        plt.xlabel("Input Parameter (x)")
        plt.ylabel("Weather Parameter (y)")
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.grid(color='gray', linestyle='--', linewidth=0.5)
        plt.legend()
        plt.show()
    except ValueError:
        print("Error: Please enter valid numeric values.")
visualize_weather_model_input()
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("m.pujitha BU22CSEN0101365")
def plot_quadratic_from_file(filename):
    try:
        with open(filename, 'r') as file:
            line = file.readline().strip()
            try:
                coefficient_a, coefficient_b, constant_c = map(float, line.split(','))
                x_values = np.linspace(-10, 10, 100)
                y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + constant_c
                plt.figure(figsize=(8, 6))
                plt.plot(
                    x_values, y_values,
                    label=f"y = {coefficient_a}x + {coefficient_b}x + {constant_c}", color='purple'
                )
                plt.title("Weather Model - Quadratic Equation from File")
                plt.xlabel("Input Parameter (x)")
                plt.ylabel("Weather Parameter (y)")
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.legend()
                plt.show()
            except ValueError:
                print(f"Error: Invalid format in file. Line content: {line}")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
plot_quadratic_from_file('quadratic_input.txt')
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("m.pujitha BU22CSEN0101365")
def plot_multiple_quadratics_from_file(filename):
    try:
        with open(filename, 'r') as file:
            plt.figure(figsize=(10, 7))
            for line in file:
                try:
                    coefficient_a, coefficient_b, constant_c = map(float, line.strip().split(','))
                    x_values = np.linspace(-10, 10, 100)
                    y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + constant_c
                    plt.plot(
                        x_values, y_values,
                        label=f"y = {coefficient_a}x + {coefficient_b}x + {constant_c}"
                    )
                except ValueError:
                    print(f"Error: Invalid format in line: {line.strip()}")
            plt.title("Weather Model Visualization - Multiple Quadratic Equations")
            plt.xlabel("Input Parameter (x)")
            plt.ylabel("Weather Parameter (y)")
            plt.axhline(0, color='black', linewidth=0.5)
            plt.axvline(0, color='black', linewidth=0.5)
            plt.grid(color='gray', linestyle='--', linewidth=0.5)
            plt.legend()
            plt.show()
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
plot_multiple_quadratics_from_file('multiple_quadratics.txt')
```
