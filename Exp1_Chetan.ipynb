```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate roots
def equation_roots(a, b, c):
    dis = b * b - 4 * a * c
    if dis > 0:
        root1 = (-b + math.sqrt(dis)) / (2 * a)
        root2 = (-b - math.sqrt(dis)) / (2 * a)
        return (root1, root2)
    elif dis == 0:
        root = -b / (2 * a)
        return (root,)
    else:
        return None  # Complex roots

# Hard-coded example
a, b, c = 1, -3, 2
roots = equation_roots(a, b, c)
print("Roots:", roots)

# Plotting the quadratic function
x = np.linspace(-10, 10, 400)
y = a * x**2 + b * x + c
plt.plot(x, y, label='Quadratic Function')
if roots:
    for root in roots:
        plt.plot(root, 0, 'ro')  # Mark roots
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate roots
def equation_roots(a, b, c):
    dis = b * b - 4 * a * c
    if dis > 0:
        root1 = (-b + math.sqrt(dis)) / (2 * a)
        root2 = (-b - math.sqrt(dis)) / (2 * a)
        return (root1, root2)
    elif dis == 0:
        root = -b / (2 * a)
        return (root,)
    else:
        return None  # Complex roots

# Function to plot the quadratic function
def plot_quadratic(a, b, c, roots):
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c
    plt.plot(x, y, label='Quadratic Function')

    if roots:
        for root in roots:
            plt.plot(root, 0, 'ro')  # Mark roots
            plt.annotate(f'Root: {root:.2f}', xy=(root, 0), xytext=(root, 5),
                         arrowprops=dict(facecolor='black', shrink=0.05))

    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.title('Quadratic Function')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()

# Main function to get user input and calculate roots
def main():
    # User-defined input for coefficients
    try:
        a = float(input("Enter coefficient a (non-zero): "))
        if a == 0:
            raise ValueError("Coefficient 'a' must be non-zero.")
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))

        # Calculate roots
        roots = equation_roots(a, b, c)
        print("Roots:", roots)

        # Plot the quadratic function
        plot_quadratic(a, b, c, roots)

    except ValueError as e:
        print(f"Input error: {e}")

if __name__ == "__main__":
    main()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate roots
def equation_roots(a, b, c):
    dis = b * b - 4 * a * c
    if dis > 0:
        root1 = (-b + math.sqrt(dis)) / (2 * a)
        root2 = (-b - math.sqrt(dis)) / (2 * a)
        return (root1, root2)
    elif dis == 0:
        root = -b / (2 * a)
        return (root,)
    else:
        return None  # Complex roots

# Function to plot the quadratic function
def plot_quadratics(coefficients):
    x = np.linspace(-10, 10, 400)
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)

    for a, b, c in coefficients:
        y = a * x**2 + b * x + c
        plt.plot(x, y, label=f'Quadratic: {a}x² + {b}x + {c}')

        # Calculate roots
        roots = equation_roots(a, b, c)
        if roots:
            for root in roots:
                plt.plot(root, 0, 'ro')  # Mark roots
                plt.annotate(f'Root: {root:.2f}', xy=(root, 0), xytext=(root, 5),
                             arrowprops=dict(facecolor='black', shrink=0.05))

    plt.title('Quadratic Functions')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

# Main function to read coefficients from a file and plot quadratics
def main():
    filename = 'coefficients1.txt'
    coefficients = []

    try:
        with open(filename, 'r') as file:
            for line in file:
                # Read coefficients from each line
                coefficients_line = line.strip().split()
                if len(coefficients_line) != 3:
                    print(f"Invalid line: {line.strip()}. Skipping.")
                    continue

                a, b, c = map(float, coefficients_line)

                if a == 0:
                    print(f"Coefficient 'a' must be non-zero. Skipping line: {line.strip()}.")
                    continue

                coefficients.append((a, b, c))

        # Plot all quadratic functions
        plot_quadratics(coefficients)

    except FileNotFoundError:
        print(f"File {filename} not found.")
    except ValueError as e:
        print(f"Value error: {e}")

if __name__ == "__main__":
    main()
```
