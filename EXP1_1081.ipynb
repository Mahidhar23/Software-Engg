```python
import matplotlib.pyplot as plt
import numpy as np

print("T.KARUNASREE BU22CSEN0101081")

def display_weather_trend():
    coefficient_a = 10
    coefficient_b = -5
    coefficient_c = 15

    x_values = np.linspace(-10, 10, 100)
    y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + coefficient_c
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values,
             label=f"y = {coefficient_a}x² + {coefficient_b}x + {coefficient_c}",
             color='blue')
    plt.title("Weather Prediction Model - Quadratic Trend")
    plt.xlabel("Input Parameter (x)")
    plt.ylabel("Weather Parameter (y)")
    plt.axhline(0, color='black', linewidth=0.5)  # X-axis
    plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()
display_weather_trend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("T.KARUNASREE BU22CSEN0101081")
def visualize_weather_model():
    try:
        coefficient_a = float(input("Enter coefficient a: "))
        coefficient_b = float(input("Enter coefficient b: "))
        coefficient_c = float(input("Enter coefficient c: "))
        x_values = np.linspace(-10, 10, 100)
        y_values = coefficient_a * (x_values ** 2) + coefficient_b * x_values + coefficient_c
        plt.figure(figsize=(8, 6))
        plt.plot(x_values, y_values,
                 label=f"y = {coefficient_a}x² + {coefficient_b}x + {coefficient_c}",
                 color='green')
        plt.title("Weather Modeling - Quadratic Equation (User Input)")
        plt.xlabel("Input Parameter (x)")
        plt.ylabel("Weather Parameter (y)")
        plt.axhline(0, color='black', linewidth=0.5)
        plt.axvline(0, color='black', linewidth=0.5)
        plt.grid(color='gray', linestyle='--', linewidth=0.5)
        plt.legend()
        plt.show()

    except ValueError:
        print("Invalid input. Please enter valid numeric values for the coefficients.")
visualize_weather_model()
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("T.KARUNASREE BU22CSEN0101081")

def plot_single_quadratic_from_file(filename):
    try:
        with open(filename, 'r') as file:
            line_data = file.readline().strip()
            try:
                coeff_a, coeff_b, coeff_c = map(float, line_data.split(','))
                x_values = np.linspace(-10, 10, 100)
                y_values = coeff_a * (x_values ** 2) + coeff_b * x_values + coeff_c

                plt.figure(figsize=(8, 6))
                plt.plot(x_values, y_values, label=f"y = {coeff_a}x² + {coeff_b}x + {coeff_c}", color='blue')
                plt.title("Weather Model - Single Quadratic Equation")
                plt.xlabel("Input Parameter (x)")
                plt.ylabel("Weather Parameter (y)")
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.legend()
                plt.show()
            except ValueError:
                print(f"Error: Invalid data format in file: {line_data}")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' could not be found.")

plot_single_quadratic_from_file('single_input1.txt')
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("T.KARUNASREE BU22CSEN0101081")

def plot_multiple_quadratics_from_file(filename):
    try:
        with open(filename, 'r') as file:
            plt.figure(figsize=(10, 7))

            for line_data in file:
                try:
                    coeff_a, coeff_b, coeff_c = map(float, line_data.strip().split(','))
                    x_values = np.linspace(-10, 10, 100)
                    y_values = coeff_a * (x_values ** 2) + coeff_b * x_values + coeff_c
                    plt.plot(x_values, y_values, label=f"y = {coeff_a}x² + {coeff_b}x + {coeff_c}")
                except ValueError:
                    print(f"Error: Invalid data format in line: {line_data.strip()}")

            plt.title("Weather Model - Multiple Quadratic Equations")
            plt.xlabel("Input Parameter (x)")
            plt.ylabel("Weather Parameter (y)")
            plt.axhline(0, color='black', linewidth=0.5)
            plt.axvline(0, color='black', linewidth=0.5)
            plt.grid(color='gray', linestyle='--', linewidth=0.5)
            plt.legend()
            plt.show()
    except FileNotFoundError:
        print(f"Error: The file '{filename}' could not be found.")

plot_multiple_quadratics_from_file('multi_input2.txt')
```
