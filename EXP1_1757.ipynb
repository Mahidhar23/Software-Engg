## Implement weather modeling* using the quadratic solution in stages: hard-coding variables keyboard input, read from a file, for a single set of input, multiple sets of inputs.

# Hard-Coding Variables

```python
import numpy as np
import matplotlib.pyplot as plt

a = 12
b = 3
c = 1.2

x_values = np.linspace(-10, 10, 100)

y_values = a * x_values**2 + b * x_values + c

plt.plot(x_values, y_values, label=r'$y = 12x^2 + 3x + 1.2$')
plt.title('Graph of the quadratic equation y = 12x^2 + 3x + 1.2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()

plt.show()

x = 1
y = a*x**2 + b*x + c
print(f"The predicted value is {y} units")
```

# Keyboard Input

```python
import numpy as np
import matplotlib.pyplot as plt

a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))
c = int(input("Enter the value of c: "))
x_input = int(input("Enter the value of x: "))

y = a * x**2 + b * x + c

print(f"The predicted value is {y} units")

x_values = np.linspace(-10, 10, 100)

y_values = a * x_values**2 + b * x_values + c

plt.plot(x_values, y_values, label=f'Quadratic equation: y = {a}x^2 + {b}x + {c}')
plt.scatter(x_input, y, color='red', label=f'Point ({x_input}, {y})')
plt.title(f'Graph of y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

with open("/content/input_single.txt", "r") as file:
    data = file.readlines()

a = int(data[0].strip())
b = int(data[1].strip())
c = int(data[2].strip())
x = 1

y = a * x**2 + b * x + c

if y:
    print(f"The model's prediction if it 'Rains' tomorrow is {y}%")
else:
    print("Some error has occurred, contact developer")

x_values = np.linspace(-10, 10, 100)

y_values = a * x_values**2 + b * x_values + c

plt.plot(x_values, y_values, label=f'Quadratic equation: y = {a}x^2 + {b}x + {c}')
plt.scatter(x, y, color='red', label=f'Point ({x}, {y})')
plt.title(f'Graph of y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

y_val = []

x_values = list(range(1, 11))

with open("input.txt", "r") as file:
    data = file.readlines()

for i in range(0, len(data), 3):
    a = int(data[i].strip())
    b = int(data[i+1].strip())
    c = int(data[i+2].strip())

    x = i // 3 + 1

    y = a * x**2 + b * x + c

    if y:
        print(f"The model's prediction if it 'Rains' tomorrow is {y}% for x = {x}")
        y_val.append(y)
    else:
        print("Some error has occurred, contact developer")

plt.plot(x_values[:len(y_val)], y_val, marker='*', label="Rain Prediction")
plt.title("Model Prediction Values")
plt.xlabel("Dataset Values (By 10s)")
plt.ylabel("Prediction Value (y)")
plt.legend()
plt.grid(True)
plt.show()
```
